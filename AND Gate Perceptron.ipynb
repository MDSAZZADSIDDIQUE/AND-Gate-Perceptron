```python
import math
import matplotlib.pyplot as plt
import random
```

```python
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
Y = [0, 0, 0, 1]
```

```python
plt.figure()
for i in range(len(Y)):
    if Y[i] == 0:
        plt.plot(X1[i], X2[i], "ro")
    else:
        plt.plot(X1[i], X2[i], "go")
plt.show()
```

```python
W = [random.random(), random.random()]
b = random.random()

print(W, b)

learning_rate = 0.02
theta = 0.5
```

```python
epoch = 0
count = 0

while count < 4:
    total_error = 0
    print(f"==========")
    print(f"Epoch: {epoch}")
    print(f"==========")
    for i in range(len(X1)):
        weighted_sum = W[0] * X1[i] + W[1] * X2[i] + b
        pred = 0
        if weighted_sum > theta:
            pred = 1
        error = Y[i] - pred
        if error != 0:
            count = 0
            total_error += math.sqrt(error**2)
            W[0] += learning_rate * error * X1[i]
            W[1] += learning_rate * error * X2[i]
            b += learning_rate * error
        else:
            count += 1
        print(f"[{X1[i]}, {X2[i]}] = {pred}")
        print(f"----------")    
    print(f"Loss: {total_error}")
    epoch += 1
```

```python
print(f"W[0] = {W[0]}")
print(f"W[1] = {W[1]}")
print(f"b = {b}")
```

```python
plt.figure()
for i in range(len(Y)):
    if Y[i] == 0:
        plt.plot(X1[i], X2[i], "ro")
    else:
        plt.plot(X1[i], X2[i], "go")
xintr = (theta - b) / W[0]
yintr = (theta - b) / W[1]
slope = -yintr/xintr
xx = plt.xlim()
yeq = []
for x in xx:
    yeq.append(slope * x + yintr)
plt.plot(xx, yeq)
plt.show()
```
